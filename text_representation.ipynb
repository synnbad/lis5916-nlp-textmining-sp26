```python
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
 
```

```python
import numpy as np

#Reduce dimensions using PCA
words = list(model.wv.index_to_key)[:50]
vectors = np.array([model.wv[word] for word in words])
pca = PCA(n_components=2)
reduced_vectors = pca.fit_transform(vectors)
```

```python
#Plot words in 2D space
plt.figure(figsize=(8,6))
for i, word in enumerate(words):
    plt.scatter(reduced_vectors[i,0], reduced_vectors[i,1])
    plt.annotate(str(word), xy=(reduced_vectors[i,0], reduced_vectors[i,1]))
plt.title("Word Embeddings Visualization using PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
 
```

```python
import re
from gensim.models import FastText
from gensim.utils import simple_preprocess
```

```python
path = 'alexander_the_great.txt'  
```

```python
with open(path, 'r', encoding='utf-8') as file:
    text = file.read()
    
```

```python
sentences_raw = re.split(r"[.!?]\s+|\n+", text)  # Split into sentences based on punctuation
```

```python
sentences_raw
```

```python
sentences = []
for s in sentences_raw:
    tokens = simple_preprocess(s,deacc=True)  # Tokenize and preprocess
    if tokens:
        sentences.append(tokens)
```

```python
print(f"Loaded {len(sentences)} sentences")
print("Example tokenized sentence:", sentences[0])
```

```python
print("n\Training FastText model...")
model = FastText(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1,   
    epochs=10)

print("Model training completed.")
```

```python
def show_similar_words(word, topn=10):
    try:
        similar_words = model.wv.most_similar(word, topn=topn)
        print(f"Words similar to '{word}':")
        for similar_word, similarity in similar_words:
            print(f"  {similar_word}: {similarity:.4f}")
    except KeyError:
        print(f"Word '{word}' not found in the vocabulary.")
```

```python
show_similar_words('alexandrian')
```

```python
show_similar_words('battle')
```

```python
def show_analogy(positive, negative=None, topn=5):
    negative = negative or []
    print(f"\nAnalogy: +{positive} - {negative}")
    try:
        results = model.wv.most_similar(positive=positive, negative=negative, topn=topn)
        for w, score in results:
            print(f"  {w}: {score:.3f}")
    except KeyError as e:
        print("Missing word in vocabulary:", e)
```

```python
show_analogy(positive=['pharaoh', 'macedonia'], negative=['persian'])
```
