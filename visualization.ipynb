```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme()
x = np.linspace(0,10,100)
y = np.power(x,2)
plt.plot(x,y)
plt.show()
```

```python
samp1   = np.random.normal(loc=0, scale=1, size=1000)
samp2   = np.random.normal(loc=2, scale=1.5, size=1000)
samp3   = np.random.normal(loc=-1, scale=0.5, size=1000)

f, ax = plt.subplots(figsize=(10, 6))
ax.boxplot((samp1, samp2, samp3))
ax.set_xticklabels(['Sample 1', 'Sample 2', 'Sample 3'])
plt.savefig("boxplot.pdf")
```

```python
data = np.random.normal(size=1000)
f, ax = plt.subplots(1,2, figsize=(6, 3))

ax[0].hist(data, bins=30, color='blue', density=True)

ax[1].hist(data, bins=30, orientation='vertical', color='red', density=True, cumulative=True)
plt.savefig("histogram.pdf")
```
